```python
from pyspark.sql.functions import *
support_ticket_bronze_df=spark.read.format('delta').load('abfss://data@storagecustomer260.dfs.core.windows.net/catalog/bronze/support_tickets/data')
support_clean_df=support_ticket_bronze_df.withColumn('category',when((col('customer_id').isNotNull())&(col('opened_date').isNotNull())&(col('ticket_status').isin('Open','Closed'))&(((col('ticket_status')=='Closed')&(col('closed_date').isNotNull()))|((col('ticket_status')=='Open')&(col('closed_date').isNull()))),'valid').otherwise('bad'))
total_count=support_clean_df.select('ticket_id').count()
distinct_count=support_clean_df.select('ticket_id').distinct().count()
if total_count==distinct_count:
    display('No Duplicates')
else:
    duplicate_df=support_clean_df.groupBy('ticket_id').agg(count('*').alias('total_count')).filter(col('total_count')>1)
    duplicate_list=duplicate_df.select('ticket_id').collect()
    duplicate_ticket_lists=[row['ticket_id'] for row in duplicate_list]
    support_clean_df=support_clean_df.withColumn('category',when(col('ticket_id').isin(duplicate_ticket_lists),'bad').otherwise(
        col('category')
    ))
support_clean_df=support_clean_df.withColumn('opened_date',to_date('opened_date','dd-MM-yyyy'))
support_clean_df=support_clean_df.withColumn('closed_date',to_date('closed_date','dd-MM-yyyy'))
display(support_clean_df)
support_clean_df.write.format('delta').partitionBy('category').mode('overwrite').option('mergeSchema',True).save('abfss://data@storagecustomer260.dfs.core.windows.net/catalog/silver/support_tickets/data')
```

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
web_activity_bronze_df=spark.read.format('delta').load('abfss://data@storagecustomer260.dfs.core.windows.net/catalog/bronze/web_activity/data')
##web_activity_clean_df=web_activity_clean_df.withColumn('last_login', expr("try_cast(last_login AS DATE)"))
web_activity_clean_df=web_activity_bronze_df.withColumn('category',when((col('customer_id').isNotNull())&(col('last_login')<=current_date())&(col('duration')>0)&(col('pages_visited')>0),'valid').otherwise('bad'))
total_count_web=web_activity_clean_df.select('activity_id').count()
total_distinct_count_web=web_activity_clean_df.select('activity_id').distinct().count()
if total_count_web==total_distinct_count_web:
    display('No Duplicates')
else:
    duplicate_web_df=web_activity_clean_df.groupBy('activity_id').agg(count('*').alias('total_count')).filter(col('total_count')>1)
    duplicate_web_rows=duplicate_web_df.select('activity_id').collect()
    duplicate_web_list=[row['activity_id'] for row in duplicate_web_rows]
    web_activity_clean_df=web_activity_clean_df.withColumn('category',when(col('activity_id').isin(duplicate_web_list),'bad').otherwise(
        col('category')
    ))
display(web_activity_clean_df)
web_activity_clean_df.write.format('delta').partitionBy('category').mode('overwrite').option('mergeSchema',True).save('abfss://data@storagecustomer260.dfs.core.windows.net/catalog/silver/web_activity/data')
```
