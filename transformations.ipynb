```python
from pyspark.sql.functions import *
transaction_silver_df=spark.read.format('delta').load('abfss://data@storagecustomer260.dfs.core.windows.net/catalog/silver/transactions/data').filter(col('category')=='valid')
customer_master_df=spark.read.format('delta').load('abfss://data@storagecustomer260.dfs.core.windows.net/catalog/silver/customer_master_data/data')
combined_df=transaction_silver_df.join(customer_master_df,transaction_silver_df.customer_id==customer_master_df.customer_id,how='inner').drop(customer_master_df.customer_id,transaction_silver_df.category,transaction_silver_df._rescued_data,transaction_silver_df.record_source,transaction_silver_df.ingestion_timestamp)
display(combined_df)
```

```python
from pyspark.sql.functions import *
rfm_df=combined_df.groupBy('customer_id').agg(max('transaction_date').alias('latest_order_date'),count('transaction_id').alias('frequency'),sum('amount').alias('monetary')).orderBy(col('customer_id'))
rfm_df=rfm_df.withColumn('recency',datediff(current_date(),rfm_df.latest_order_date)).drop('latest_order_date')
rfm_df=rfm_df.withColumn('churn_status',when((col('recency')<=30),'Active')\
    .otherwise(when(((col('recency') >= 31) & (col('recency') <= 90)),'At_Risk')\
    .otherwise('Churn_Likely'))
    ).withColumn("classification_timestamp", current_timestamp())
rfm_df.write.format('delta').mode('overwrite').save('abfss://data@storagecustomer260.dfs.core.windows.net/catalog/silver/customer_classification/data')
```

```python
from pyspark.sql.functions import *

rfm_silver_df = spark.read.format('delta').load(
    'abfss://data@storagecustomer260.dfs.core.windows.net/catalog/silver/customer_classification/data'
)

rfm_modify_df = rfm_silver_df.withColumn(
    'customer_type',
    when((col('recency') <= 30) & (col('frequency') >= 3) & (col('Monetary') >= 5000), 'Champions')
    .otherwise(
        when((col('recency') <= 60) & (col('frequency') >= 3), 'Loyal_customers')
        .otherwise(
            when((col('Monetary') >= 5000) & (col('frequency') < 3), 'Big Spenders')
            .otherwise(
                when((col('recency') > 90), 'Lost')
                .otherwise(
                    when(
                        ((col('recency') > 60) & (col('recency') <= 90) & (col('frequency') >= 2)) |
                        (col('Monetary') >= 2000),
                        'At Risk'
                    )
                    .otherwise('Lost')
                )
            )
        )
    )
)

rfm_modify_df.write.format('delta').mode('overwrite').save('abfss://data@storagecustomer260.dfs.core.windows.net/catalog/silver/customer_classification_segmentation/data')
```

```python
from pyspark.sql.functions import *
support_ticket_silver_df=spark.read.format('delta').load('abfss://data@storagecustomer260.dfs.core.windows.net/catalog/silver/support_tickets/data').filter(col('category')=='valid')
closed_tickets_df=support_ticket_silver_df.filter(col('ticket_status')=='Closed')
open_tickets_df=support_ticket_silver_df.filter(col('ticket_status')=='Open')
closed_tickets_df=closed_tickets_df.withColumn('resolution_time',datediff(col('closed_date'),col('opened_date')))
closed_tickets_metrics_df=closed_tickets_df.groupBy('customer_id').agg(count('ticket_id').alias('total_closed_tickets'),round(avg("resolution_time"),2).alias('avg_resolution_time'))
open_tickets_support_df=open_tickets_df.groupBy('customer_id').agg(count('ticket_id').alias('total_open_tickets'))
total_tickets_count_df=support_ticket_silver_df.groupBy('customer_id').agg(count('ticket_id').alias('total_ticket_count'))
support_ticket_metrics_df=closed_tickets_metrics_df.join(open_tickets_support_df,closed_tickets_metrics_df.customer_id==open_tickets_support_df.customer_id,how='left').drop(open_tickets_support_df.customer_id)
support_ticket_metrics_df=support_ticket_metrics_df.join(total_tickets_count_df,support_ticket_metrics_df.customer_id==total_tickets_count_df.customer_id,how='left').drop(total_tickets_count_df.customer_id)
support_ticket_metrics_df=support_ticket_metrics_df.withColumn('segmentation',when(col('total_ticket_count')<=1,'Low Complaints')\
    .otherwise(when((col('total_ticket_count')>=2)&(col('total_ticket_count')<=4),'Moderate Complaints')\
        .otherwise(when((col('total_open_tickets')>=3)&(col('total_ticket_count')>=3),'Pending Risk')\
            .otherwise(when(col('total_ticket_count')>=5,'High Complaints')\
                .otherwise('Neutral users')
                ))
    ))
support_ticket_metrics_df.write.format('delta').mode('overwrite').option('mergeSchema',True).save('abfss://data@storagecustomer260.dfs.core.windows.net/catalog/silver/support_ticket_metrics/data')
```

```python
from pyspark.sql.functions import *
web_activity_silver_df=spark.read.format('delta').load('abfss://data@storagecustomer260.dfs.core.windows.net/catalog/silver/web_activity/data').filter(col('category')=='valid')
web_activity_metrics_df=web_activity_silver_df.groupBy('customer_id').agg(max('last_login').alias('latest_login_date'),avg('duration').alias('average_duration_time'),sum('pages_visited').alias('total_pages_visited'))
web_activity_metrics_df=web_activity_metrics_df.withColumn('days_since_last_login',datediff(current_date(),col('latest_login_date')))
web_activity_metrics_df=web_activity_metrics_df.withColumn('engagement_level',when((col('total_pages_visited')>=20)&(col('days_since_last_login')<=15),'Highly Engaged')\
.otherwise(when((col('days_since_last_login')<=10)&(col('average_duration_time')<5),'Browsers')\
    .otherwise(when(((col('days_since_last_login'))>30)&(col('total_pages_visited')<5),'Silent Leavers')\
        .otherwise(when(col('days_since_last_login')>=90,'Dormant Users')\
            .otherwise('Neutral Users')))
    )    
)
web_activity_metrics_df.write.format('delta').mode('overwrite').option('mergeSchema',True).save('abfss://data@storagecustomer260.dfs.core.windows.net/catalog/silver/web_activity_metrics/data')
```
