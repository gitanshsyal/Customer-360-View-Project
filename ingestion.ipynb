```python
transactions_raw_Df=spark.readStream.format('cloudFiles')\
.option('cloudFiles.format','csv')\
.option('cloudFiles.schemaLocation','abfss://data@storagecustomer260.dfs.core.windows.net/catalog/bronze/transactions/schema')\
.option('cloudFiles.schemaEvolutionMode','addNewColumns')\
.load('abfss://data@storagecustomer260.dfs.core.windows.net/catalog/raw/main_files/transactions')
```

```python
transactions_raw_Df.writeStream.format('delta').option('mode','append').option('checkpointLocation','abfss://data@storagecustomer260.dfs.core.windows.net/catalog/bronze/transactions/checkpoint').option('mergeSchema',True).trigger(availableNow=True).start('abfss://data@storagecustomer260.dfs.core.windows.net/catalog/bronze/transactions/data')
```

```python
customer_schema='Customer_id int,first_name string,last_name string,email string,phone string,address string'
customer_raw_df=spark.read.format('csv')\
    .schema(customer_schema)\
    .option('header',True)\
    .load('abfss://data@storagecustomer260.dfs.core.windows.net/catalog/raw/main_files/customer_updates/')
```

```python
customer_raw_df.write.format('delta').mode('overwrite').option('mergeSchema',True).save('abfss://data@storagecustomer260.dfs.core.windows.net/catalog/bronze/customer_updates/data')
```
